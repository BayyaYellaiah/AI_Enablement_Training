**Load Image**

```python
from PIL import Image
import matplotlib.pyplot as plt

traffic_img = Image.open("images/image_2.jpeg")

# Display the image using Matplotlib
plt.figure(figsize=(6, 4))
plt.imshow(traffic_img)
plt.axis('off')  # Hide the axes
plt.show()
```

```python
# Import pipeline
from transformers import pipeline

# Specify the inference task
object_detection = pipeline(
    task="object-detection",
    model="facebook/detr-resnet-50",
)
```

```python
# Pass the input to the pipeline
detections = object_detection(traffic_img)

detections[:4]
```

```python
from collections import defaultdict

res = defaultdict(list)
for detection in detections:
    res[detection['label']].append(detection['score'])

message = "We have" 
i=0;
for k,v in res.items():
    if(i==len(res.items())-1):
        message = f"{message} {len(v)} {k}'s "
    else:
        message = f"{message} {len(v)} {k}'s, "
    i+=1
message +="in the given image"

print(message)
```

```python
narrator = pipeline(
    task="text-to-speech",
    model="suno/bark-small",
)
```

```python
narrated_text = narrator(message)
from IPython.display import Audio as IPythonAudio

IPythonAudio(
    narrated_text["audio"][0],
    rate=narrated_text["sampling_rate"]
)
```
